```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.display import Image, display

from utils.common import name_to_alias
from db.database import load_time_series_as_numpy
from dev import load_and_split
```

```python
torch.manual_seed(1337)
np.random.seed(1337)
```

```python
DATASET_NAME = "Quick Harmon"
DATASET_ALIAS = name_to_alias(DATASET_NAME)

BATCH_SIZE = 20
TRAIN_SPLIT = 0.3
```

```python
dataset = load_time_series_as_numpy(DATASET_NAME)
```

```python
x_train, y_train, x_test, y_test = load_and_split(dataset, ratio=TRAIN_SPLIT, batch_size=BATCH_SIZE)
```
